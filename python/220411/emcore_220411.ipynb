### ******집계*****
##### group by
* max, min, median, meadn, std, var

```python
# 집계 하고싶은 대상,  그 다음알고싶은 것
kbo_df2.groupby(['Team'])[['연봉(만원)']].mean().sort_values(by=['연봉(만원)'])
```

```python
#중간값
kbo_df2.groupby(['Team'])[['연봉(만원)']].median().sort_values(by=['연봉(만원)'])
```

```python
#최고
kbo_df2.groupby(['Team'])[['연봉(만원)']].max().sort_values(by=['연봉(만원)'])
```

```python
#표준편차
kbo_df2.groupby(['Team'])[['연봉(만원)']].std().sort_values(by=['연봉(만원)'])
```

* max, min, median, meadn, std, var 익러 원샷 6킬 하는법

```python
kbo_df2.groupby(['Team'])[['연봉(만원)']].agg(['max', 'min', 'std', 'var', 'median', 'mean'])
```
